```python
def save_dicts_to_jsonl(file_path, list_of_dicts):
    with open(file_path, 'w') as f:
        for dictionary in list_of_dicts:
            json_str = json.dumps(dictionary)
            f.write(json_str + '\n')

def generate_data(fake, num_records):
    # Calculate the date range for the past 6 months
    end_date = datetime.now()
    start_date = end_date - timedelta(days=1825)

    patients = []
    medical_histories = []
    allergies = []
    medications = []
    appointments = []
    billings = []

    for _ in range(num_records):
        # Generate patient data
        patient = {
            "patient_id": _,
            "first_name": fake.first_name(),
            "last_name": fake.last_name(),
            "date_of_birth": fake.date_of_birth(minimum_age=18, maximum_age=90).isoformat(),
            "address": fake.address(),
            "phone": fake.phone_number(),
            "email": fake.email(),
            "blood_type": fake.random_element(elements=('A+', 'A-', 'B+', 'B-', 'AB+', 'AB-', 'O+', 'O-')),
        }
        patients.append(patient)

        # Generate allergy
        allergy = {
            "patient_id": patient["patient_id"],
            "allergen": fake.word(ext_word_list=['Peanuts', 'Penicillin', 'Pollen', 'Dust', 'None']),
            "severity": fake.random_element(elements=('Mild', 'Moderate', 'Severe'))
        }
        allergies.append(allergy)

        # Generate medication
        medication = {
            "patient_id": patient["patient_id"],
            "medication_name": fake.word(ext_word_list=['Aspirin', 'Lisinopril', 'Metformin', 'Atorvastatin']),
            "dosage": f"{random.randint(1, 1000)} mg",
            "frequency": fake.random_element(elements=('Once daily', 'Twice daily', 'As needed')),
            "start_date": fake.date_between(start_date=start_date, end_date=end_date).isoformat(),
            "end_date": fake.date_between(start_date=start_date, end_date=end_date).isoformat()
        }
        medications.append(medication)

        # Generate multiple appointments and billing records per patient
        num_appointments = random.randint(1, 10)
        for j in range(num_appointments):
            condition = fake.word(ext_word_list=list(condition_notes.keys()))  # Choose a condition
            notes = fake.sentence(ext_word_list=condition_notes[condition])  # Pick notes relevant to the condition

            appointment_id = f"{patient['patient_id']}-{j}"  # Unique appointment ID
            appointment = {
                "appointment_id": appointment_id,
                "patient_id": patient["patient_id"],
                "doctor_id": random.randint(1, 100),
                "appointment_date": fake.date_time_between(start_date=start_date, end_date=end_date).isoformat(),
                "condition": condition,
                "notes": notes
            }
            appointments.append(appointment)

            # Generate billing linked to each appointment
            billing = {
                "patient_id": patient["patient_id"],
                "appointment_id": appointment_id,
                "amount": round(random.uniform(100, 100000), 2),
                "insurance_claim_id": fake.uuid4(),
                "status": fake.random_element(elements=('Pending', 'Paid', 'Denied')),
                "payment_date": fake.date_between(start_date=start_date, end_date=end_date).isoformat()
            }
            billings.append(billing)

    return patients, medical_histories, allergies, medications, appointments, billings



def main():
    fake = Faker()
    num_records = 1000000
    patients, medical_histories, allergies, medications, appointments, billings = generate_data(fake, num_records)

    # Save data to JSON files
    save_dicts_to_jsonl('patients.json', patients)
    save_dicts_to_jsonl('medical_histories.json', medical_histories)
    save_dicts_to_jsonl('allergies.json', allergies)
    save_dicts_to_jsonl('medications.json', medications)
    save_dicts_to_jsonl('appointments.json', appointments)
    save_dicts_to_jsonl('billings.json', billings)
    # upload_to_s3('health-tech-demo',patients, 'patients.json')
    # upload_to_s3('health-tech-demo',medical_histories, 'medical_histories.json')
    # upload_to_s3('health-tech-demo',allergies, 'allergies.json')
    # upload_to_s3('health-tech-demo',medications, 'medications.json')
    # upload_to_s3('health-tech-demo',appointments, 'appointments.json')
    # upload_to_s3('health-tech-demo',billings, 'billings.json')


    print("Data generation complete.")

if __name__ == "__main__":
    main()
```
